```python
import numpy as np
import pandas as pd
from typing import Tuple
import tensorflow as tf
from tensorflow.keras.utils import plot_model
from sklearn.svm import SVC
from sklearn import svm
import warnings
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt 
from copy import deepcopy

warnings.filterwarnings('ignore')
```

```python
# Load the data and label
def load_data(filename: str) -> Tuple[np.ndarray, np.ndarray]:
    data = np.load(f'{filename}')
    return data['data'], data['labels']

# Split the data and labels into training and testing sets

train_data, train_labels = load_data('train_data_SYN.npz')

print(train_data.shape)
print(np.unique(train_labels, return_counts=True))

test_data, test_labels = load_data('test_data_SYN.npz')

print(test_data.shape)
print(np.unique(test_labels, return_counts=True))    
```

```python
train_data = pd.DataFrame(train_data)
train_labels = pd.DataFrame(train_labels)
test_data = pd.DataFrame(test_data)
test_labels = pd.DataFrame(test_labels)

train_0 = train_data.loc[train_labels[0] == 0]
train_data_0 = train_data.iloc[train_0.index]
train_data_0 = train_data_0[:100]

train_1 = train_data.loc[train_labels[0] == 1]
train_data_1 = train_data.iloc[train_1.index]
train_data_1 = train_data_1[:100]

train_2 = train_data.loc[train_labels[0] == 2]
train_data_2 = train_data.iloc[train_2.index]
train_data_2 = train_data_2[:100]

train_3 = train_data.loc[train_labels[0] == 3]
train_data_3 = train_data.iloc[train_3.index]
train_data_3 = train_data_3[:100]

train_4 = train_data.loc[train_labels[0] == 4]
train_data_4 = train_data.iloc[train_4.index]
train_data_4 = train_data_4[:100]

train_5 = train_data.loc[train_labels[0] == 5]
train_data_5 = train_data.iloc[train_5.index]
train_data_5 = train_data_5[:100]

train_6 = train_data.loc[train_labels[0] == 6]
train_data_6 = train_data.iloc[train_6.index]
train_data_6 = train_data_6[:100]

train_7 = train_data.loc[train_labels[0] == 7]
train_data_7 = train_data.iloc[train_7.index]
train_data_7 = train_data_7[:100]

train_8 = train_data.loc[train_labels[0] == 8]
train_data_8 = train_data.iloc[train_8.index]
train_data_8 = train_data_8[:100]

train_9 = train_data.loc[train_labels[0] == 9]
train_data_9 = train_data.iloc[train_9.index]
train_data_9 = train_data_9[:100]

new_train_data = pd.concat([train_data_0, train_data_1, train_data_2, train_data_3, train_data_4,
                           train_data_5, train_data_6, train_data_7, train_data_8, train_data_9])
new_train_labels = train_labels.iloc[new_train_data.index]
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
degrees = [2, 3, 4]
gammas = ['scale', 'auto']
```

```python
best_setup = None
best_accuracy = 0
for kernel in kernels:
    for degree in degrees:
        for gamma in gammas:
            # new_train_data_cpy = deepcopy(new_train_data)
            # new_train_labels_cpy = deepcopy(new_train_labels)
            #Create a svm Classifier
            clf = svm.SVC(C=1.0, kernel=kernel, degree=degree, gamma=gamma, coef0=0.0, shrinking=True,
                            probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1,
                            decision_function_shape='ovr', break_ties=False, random_state=40)
            #Train the model using the training sets
            clf.fit(new_train_data, new_train_labels)
            
            #Predict the response for test dataset
            y_pred = clf.predict(test_data)
            score = accuracy_score(y_pred, test_labels)
            print(f'we get {round(score * 100, 2)}% accuracy with {kernel} kernel and degree = {degree} and gamma = {gamma}')
            cm = confusion_matrix(test_labels, clf.predict(test_data), labels=clf.classes_)
            print(cm)
            if score > best_accuracy:
                best_accuracy = score
                best_setup = [kernel, degree, gamma]
```

```python
print(f'The best accuarcy = {round(best_accuracy * 100, 2)} belongs to this setup: {best_setup}')
clf = svm.SVC(C=1.0, kernel=best_setup[0], degree=best_setup[1], gamma=best_setup[2], coef0=0.0, shrinking=True,
                            probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1,
                            decision_function_shape='ovr', break_ties=False, random_state=40)

#Train the model using the training sets
clf.fit(train_data, train_labels)
#Predict the response for test dataset
y_pred = clf.predict(test_data)
score = accuracy_score(y_pred, test_labels)
print(f'accuracy with whole of data is {score}')
cm = confusion_matrix(test_labels, clf.predict(test_data), labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
input_tensor = tf.keras.Input(shape=(1024,))

hidden_size = 32
reg_l1_param = 10e-5

hidden_layer_1 = tf.keras.layers.Dense(units=hidden_size, activation=tf.nn.relu, 
                                       name = 'hidden_layer',
                                       activity_regularizer=tf.keras.regularizers.l1(reg_l1_param))(input_tensor)

output_layer = tf.keras.layers.Dense(units=10, activation=tf.nn.softmax, name = 'classification_layer')(hidden_layer_1)

model = tf.keras.Model(inputs=input_tensor, outputs=output_layer)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_data, train_labels, epochs=10, batch_size=32, validation_data=(test_data, test_labels))
test_loss, test_acc = model.evaluate(test_data, test_labels)

print('reg_l1_param:', reg_l1_param, ' hidden_size:', hidden_size, ' Test accuracy:', test_acc)

# Plot the model architecture
plot_model(model, to_file='model.png', show_shapes=True)
```

```python
def get_top_features(model, real_indexes):
    first_layer_weights = model.layers[1].get_weights()[0]
    first_layer_weights
    weights_mean = {}
    for i in range(first_layer_weights.shape[0]):
        weights_mean[i] = first_layer_weights[i].mean()

    sorted_weights_mean = {k: v for k, v in sorted(weights_mean.items(), key=lambda item: -item[1])}
    top_features = []
    for k, v in sorted_weights_mean.items():
        if v > 0.005:
            top_features.append(k)
    print(f'{len(top_features)} features selected.')
    new_real_indexes = []
    for feature in top_features:
        new_real_indexes.append(real_indexes[feature])
    return new_real_indexes, top_features


def fit_model(top_features):
    input_tensor = tf.keras.Input(shape=(len(top_features),))

    hidden_layer_1 = tf.keras.layers.Dense(units=best_setup[1], activation=tf.nn.relu, 
                                           name = 'hidden_layer',
                                           activity_regularizer=tf.keras.regularizers.l1(best_setup[0]))(input_tensor)

    output_layer = tf.keras.layers.Dense(units=10, activation=tf.nn.softmax, name = 'classification_layer')(hidden_layer_1)

    model = tf.keras.Model(inputs=input_tensor, outputs=output_layer)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    history = model.fit(new_train_data, train_labels, epochs=10, batch_size=32, validation_data=(new_test_data, test_labels), verbose=0)
    return model

    
real_indexes, top_features = get_top_features(model, [x for x in range(1024)])
best_score = 0
best_features = None

while len(top_features) > 10:
    new_train_data = train_data[top_features]
    new_test_data = test_data[top_features]
    model = fit_model(top_features)
    real_indexes, top_features = get_top_features(model, real_indexes)
    
    if len(top_features) < 100:
        new_train_data = train_data[top_features]
        new_test_data = test_data[top_features]
        clf = svm.SVC(C=1.0, kernel='linear', degree=2, gamma='scale', coef0=0.0, shrinking=True,
                            probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1,
                            decision_function_shape='ovr', break_ties=False, random_state=40)

        #Train the model using the training sets
        clf.fit(new_train_data, train_labels)
        #Predict the response for test dataset
        y_pred = clf.predict(new_test_data)
        accuracy = accuracy_score(y_pred, test_labels)
        score = accuracy - 0.00075 * len(top_features)
        print(f'score with {len(top_features)} features is {score}')
        if score > best_score:
            best_features = real_indexes
            best_score = score


print(f'final features are: {best_features} with score = {best_score}')
```
